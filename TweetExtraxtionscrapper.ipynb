# Getting 40K tweets at one go 

```python
import nest_asyncio
import twint
nest_asyncio.apply()
config2 = twint.Config()
config2.Search = "#MortalKombatMovie"

#configuration
config2.Lang = "en"
config2.Limit = 40000
config2.Store_json = True
config2.Output = "custom.json"
twint.run.Search(config2)
```

# Converting Json file tweets to panda datyagrame

```python
import pandas as pd
data = pd.read_json("custom.json", lines=True)
data.columns
```
